```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_revenues.csv')
```

```python
df.head()
```

```python
df['revenue_in_mln'] = df['revenue'].apply(lambda x: x/1000000)
```

```python
df.head(2)
```

```python
df.revenue_in_mln.describe()
```

```python
_, mean, std, *_ = df.revenue_in_mln.describe()
```

```python
mean, std
```

```python
def get_z_score(value, mean, std):
    return (value-mean)/std
```

```python
df['z_score'] = df.revenue_in_mln.apply(lambda x: get_z_score(x, mean, std))
```

```python
df.head(2)
```

```python
df[df.z_score > 3]
```

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s- median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_in_mln)
median = np.median(df.revenue_in_mln)
MAD, median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
df['modified_z_score'] = get_modified_z_score(df.revenue_in_mln, median, MAD)
```

```python
df.head(2)
```

```python
df[df.modified_z_score > 3.5] #display the ouliers
```
